```powershell
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```powershell
# Step 1: reading precipitatoin data
rain = pd.read_csv(r"C:\Users\ADMIN\Desktop\rain.csv")

# Step 2: 
precipitation_data = rain['rain'].values

```

```powershell
# Step 3: Compute sample moments

mean = np.mean(precipitation_data)
variance = np.var(precipitation_data)
```

```powershell
# Step 4: Parameter estimation using method of moments

alpha = mean**2 / variance
beta = mean / variance
```

```powershell
# alternative way to find parametr Fit the data

# fit_alpha, fit_loc, fit_beta = stats.gamma.fit(precipitation_data)
```

```powershell
# Generate gamma distribution using estimated parameters

gamma_distribution = stats.gamma(alpha, loc=0, scale=beta)
```

```powershell
# Plot histogram of observed data and gamma distribution PDF
plt.hist(precipitation_data, bins=30, density=True, alpha=0.7, label='Observed Data')
x = np.linspace(0, np.max(precipitation_data), 100)
plt.plot(x, gamma_distribution.pdf(x), 'r', label='Gamma Distribution')

plt.xlabel('Precipitation')
plt.ylabel('Density')
plt.title('Fitting Precipitation Data to Gamma Distribution\n ( Monthly Precipitation from 1874 to 2022 for Devon)')
plt.legend()
plt.savefig('output_plot.png', format='png')
plt.show()
```
